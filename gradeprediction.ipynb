```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Training Data
data = {
    'MST20': [19, 18, 16, 14, 12, 10, 8, 5],
    'EndSem60': [58, 54, 48, 42, 35, 28, 22, 15],
    'Attendance20': [20, 19, 17, 15, 13, 11, 9, 6],
    'Grade': [3, 3, 2, 2, 1, 1, 0, 0]  # A=3, B=2, C=1, D=0
}

df = pd.DataFrame(data)

print("Training Data")
print(df)

# Features and Target
X = df[['MST20', 'EndSem60', 'Attendance20']].values
y = df['Grade'].values

# Build Model
model = keras.Sequential([
    keras.Input(shape=(3,)),
    layers.Dense(12, activation='relu'),
    layers.Dense(8, activation='relu'),
    layers.Dense(4, activation='softmax')
])

# Compile Model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Train Model
model.fit(X, y, epochs=300, verbose=0)

# User Input
m1 = float(input("Enter Mid Sem 1 Marks (0-30): "))
m2 = float(input("Enter Mid Sem 2 Marks (0-30): "))
endsem = float(input("Enter End Sem Marks (0-60): "))
attendance_percent = float(input("Enter Attendance Percentage (0-100): "))

# Validation
if m1 < 0 or m1 > 30:
    print("Invalid Mid Sem 1 Marks!")
elif m2 < 0 or m2 > 30:
    print("Invalid Mid Sem 2 Marks!")
elif endsem < 0 or endsem > 60:
    print("Invalid End Sem Marks!")
elif attendance_percent < 0 or attendance_percent > 100:
    print("Invalid Attendance Percentage!")
else:

    # Convert MST to 20 marks
    mst20 = ((m1 + m2) / 60) * 20

    # Convert Attendance % to 20 marks
    attendance20 = (attendance_percent / 100) * 20

    # Prediction
    pred = model.predict(
        np.array([[mst20, endsem, attendance20]]),
        verbose=0
    )

    grade_index = np.argmax(pred)

    grades = {
        3: "A",
        2: "B",
        1: "C",
        0: "D"
    }

    total_marks = mst20 + endsem + attendance20

    print("\n===== RESULT =====")
    print("MST (out of 20)       :", round(mst20, 2))
    print("End Sem (out of 60)   :", endsem)
    print("Attendance (out of 20):", round(attendance20, 2))
    print("Total Marks (out of 100):", round(total_marks, 2))
    print("Predicted Grade:", grades[grade_index])
```
